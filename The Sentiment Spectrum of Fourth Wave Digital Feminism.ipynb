## Phase 3(1) - Sentiment Analysis of Captions

```python
captions_df = pd.DataFrame(df[['caption']])
captions_df.to_csv('captions.csv')
```

```python
for caption in captions_df.iterrows():
    #print(caption)
    analysis = TextBlob(str(caption))
    print(analysis.sentiment)
```

```python
f = open('sentiment1.csv','w')
f.write('Polarity,Subjectivity')
f.write('\n')

for caption in captions_df.iterrows():
    #print(caption)
    analysis = TextBlob(str(caption))
    print(analysis.sentiment)
    f.write(str(analysis.sentiment))
    f.write('\n')

f.close()
```

```python
sentiment_series1 = pd.read_csv('sentiment1.csv')

for pol in range(0,len(sentiment_series1['Polarity'])):
             sentiment_series1['Polarity'][pol] = str(sentiment_series1['Polarity'][pol]).strip('Sentiment(polarity=')
        
table = str.maketrans(dict.fromkeys('subjectivity=)'))

for sub in range(0,len(sentiment_series1['Subjectivity'])):
             sentiment_series1['Subjectivity'][sub] = str(sentiment_series1['Subjectivity'][sub]).translate(table)
             
sentiment_series1
#sentiment_series.to_csv('sample.csv')
```

## Phase 3(2) Sentiment Analysis of Captions

```python
sid = SentimentIntensityAnalyzer()
f = open('sentiment2.csv','w')
f.write('Negative,Neutral,Positive,Compound')
f.write('\n')

for caption in captions_df.iterrows():
    print(sid.polarity_scores(str(caption)))
    f.write(str(sid.polarity_scores(str(caption))))
    f.write('\n')
    
f.close()
```

```python
sentiment_series2 = pd.read_csv('sentiment2.csv')

for neg in range(0,len(sentiment_series2['Negative'])):
             sentiment_series2['Negative'][neg] = str(sentiment_series2['Negative'][neg]).strip('{\'neg\':')
        
for neu in range(0,len(sentiment_series2['Neutral'])):
             sentiment_series2['Neutral'][neu] = str(sentiment_series2['Neutral'][neu]).strip(' \'neu\':')
        
for pos in range(0,len(sentiment_series2['Positive'])):
             sentiment_series2['Positive'][pos] = str(sentiment_series2['Positive'][pos]).strip(' \'pos\':')
        
table = str.maketrans(dict.fromkeys('\'compound\':}'))

for comp in range(0,len(sentiment_series2['Compound'])):
            sentiment_series2['Compound'][comp] = str(sentiment_series2['Compound'][comp]).translate(table)
             
sentiment_series2
```

## Merging two Sentiment Phases

```python
final_sentiment = pd.merge(sentiment_series1.assign(A=1), sentiment_series2.assign(A=1), on='A').drop('A', 1)
final_sentiment.to_csv('love_final_sentiment.csv')
final_sentiment
```
